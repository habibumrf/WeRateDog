Image predictions for dogs(tsv file), we will download it programmatically as a URL by using request library.

```python
# Downloading tweet image predictions TSV using the Requests library and writing it to image_predictions.tsv
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
response = requests.get(url)
with open('image_predictioins.tsv', mode='wb') as file:
    file.write(response.content)

# Reading the tweet image-predictions-3.tsv file

image_df = pd.read_csv('image-predictions-3.tsv', sep = '\t')
```

<a id='api'> </a>
## c. Twitter API
### Using the Tweepy library to query additional data via the Twitter API (tweet_json.txt) 


```python
#Read tweet_json.txt file, line by line, and create a list of dictionaries:

twitter_api = []
with open ('tweet-json', 'r') as file:
    for line in file:
        tweet = json.loads(line)
        
        tweet_id = tweet['id']
        retweet_count = tweet['retweet_count']
        favorite_count = tweet['favorite_count']
        
        twitter_api.append({'tweet_id' : tweet_id,
                         'retweet_count' : retweet_count,
                         'favorite_count':favorite_count})
```

```python
#Read the .json file in twitter_api dataframe:
twitter_api = pd.DataFrame(twitter_api)
twitter_api.head()
```

## Data Cleaning
In this section, we will clean **all** of the issues you documented while assessing. 

we will Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
# Make copies of original pieces of data
archive_copy = df.copy()
image_copy = image_df.copy()
api_copy = twitter_api.copy()
```

### Quality Issue #1: 

#### Define:
We will be dropping "in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id & retweeted_status_timestamp" after filtering them to only tweets and keep only NaN values in `archive_copy`

#### Code

```python
#Create a loop to filter for NaN values and then drop the columns
list = ['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id',
        'retweeted_status_user_id', 'retweeted_status_timestamp']

for i in list:
    archive_copy = archive_copy[archive_copy[i].isnull()]
    archive_copy = archive_copy.drop([i], axis = 1)
```

## Visualizations

### 1. Distributions of numeric columns.

```python
#Discover numerical columns in easier way
df_numeric = df[['retweet_count', 'favorite_count', 'dog_rating', 'img_num', 'conf_1', 'conf_2', 'conf_3']]
df_numeric.hist(figsize = (12,12));
```

- Most numeric variables have right skewed distribution, that means that the least values have occurred the most.
- We have a good percentage of high prediction cases (> 0.8).
- Most of dog ratings fall between 0 and 200, and we have extreem values reached >17500.

### 2. Relationships & Correlation Coefficient between numeric variables.

```python
#Plot a heatmap to show the correlation between the numeric variables, and use a diverging palette.
sns.heatmap(df_numeric.corr(), annot = True, fmt = '.2f', cmap = 'coolwarm', center = 0);
```

- The only relationship we can mention here, is the strong positive correlation between the retweets count & favorite count. and that makes sense.
- And we can say that the more images the tweet has, the more likely to get retweets & favorites.

### 3. Do we have good 'results' about our predictions?

```python
#Percentages of true & false in result
results = df.result_1.value_counts() + df.result_2.value_counts() + df.result_3.value_counts()
plt.pie(results, labels=results.index, autopct='%1.1f%%', startangle=90,
         pctdistance=0.85, counterclock = False, labeldistance=1.1)
plt.axis('square')
plt.title('Prediction Results');
```

- That's a good percentage, almost 75% of the results are True
